```python
# Testing Work
from sklearn.preprocessing import StandardScaler
class CustomScaler:
  def __init__(self,dataframe,features):
    self.dataframe= dataframe
    self.features = features

  def fit_transform(self):
    mean_and_variance= []
    for feature in self.features:
      unscaled_feature = self.dataframe[feature].to_numpy().reshape(len(self.dataframe[feature]), 1)
      my_scaler = StandardScaler()
      my_scaler.fit(unscaled_feature)
      params= my_scaler.fit(unscaled_feature)
      mean_and_variance.append(params)

      scaled_feature = my_scaler.transform(unscaled_feature)
      self.dataframe[feature] = scaled_feature
    self.parameters = mean_and_variance
    return self.dataframe
```

Creating helping function------------------------------------------

```python
def fit_transform(dataframe, features):
  for feature in features:
    unscaled_feature = dataframe[feature].to_numpy().reshape(len(dataframe[feature]),1)
    my_scaler = StandardScaler()
    my_scaler.fit(unscaled_feature)
    scaled_feature = my_scaler.transform(unscaled_feature)
    dataframe[feature] = scaled_feature
  return dataframe
```
